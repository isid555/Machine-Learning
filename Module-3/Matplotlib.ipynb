Matplotlib plots graphs

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#plotting

plt.plot(x)
plt.plot(y)
plt.plot(z)
```

```python
plt.plot(x,y)
plt.title('Sine')
plt.xlabel('range')
plt.ylabel('Values')
plt.show()
```

```python
plt.plot(x,y,'r--')
```

```python
# 3 D scatter plot
ax = plt.axes(projection = '3d')
z  =np.random.randint(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,cmap =  'red')
plt.show()
```
