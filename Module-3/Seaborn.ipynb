```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#total bill vs tips dataset
tips = sns.load_dataset('tips')
```

```python
tips.tail()
```

```python
#visualise the data
# sns.replot(data, xaxis, yaxis)\\
sns.relplot(data=tips , x='total_bill' ,y='tip', col='time',hue='smoker' ,style='smoker' )

```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.scatterplot(data=iris ,x='sepal_length' , y='petal_length' ,hue='species' )
```

```python
#titanic dataset
titanic  = sns.load_dataset('titanic')
```

```python
diab = pd.read_csv('diabetes.csv')
```

```python
diab.head()
```

```python
sns.displot(diab['BloodPressure'])
```

```python
correlation = diab.corr()
```

Heat maps


```python
plt.figure(figsize=(10,10))
sns.heatmap(data=correlation , cbar=True,square=True , fmt='1f' ,annot=True  ,annot_kws={'size': 8},cmap = 'Blues')
```
